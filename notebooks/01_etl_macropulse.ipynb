```python
# ETL MacroPulse Brasil (BCB + IBGE + ANP)

Notebook demonstrativo do pipeline **Bronze/Silver/Gold** usando os módulos em `src/`.

O que este notebook faz:
- Executa o pipeline (`src/main.py`)
- Mostra o `summary.md` gerado
- Consulta tabelas no DuckDB (`data/macropulse.duckdb`)
```

```python
#Setup (paths e imports)

import os
import sys
from pathlib import Path

# Garante import do /src a partir do notebook
ROOT = Path.cwd().parent  # notebooks/ -> raiz do projeto
SRC = ROOT / "src"
sys.path.append(str(SRC))

print("ROOT:", ROOT)
print("SRC:", SRC)
```

```python
#Executar o pipeline (chamando o main)

import subprocess

subprocess.run(["python", str(ROOT / "src" / "main.py")], check=True)
```

```python
#Ler o summary gerado

summary_path = ROOT / "data" / "gold" / "summary.md"
print(summary_path.read_text(encoding="utf-8"))
```

```python
#Consultas no DuckDB

import duckdb

db_path = ROOT / "data" / "macropulse.duckdb"
con = duckdb.connect(str(db_path))

con.execute("SHOW TABLES;").df()
```

```python
#Amostra Silver (BCB)

con.execute("""
    SELECT series_id, series_name, date, value
    FROM silver_bcb_sgs
    ORDER BY date DESC
    LIMIT 10;
""").df()
```

```python
#Amostra Gold (ANP mensal)
con.execute("""
    SELECT uf_sigla, product, month, avg_price
    FROM gold_anp_monthly
    ORDER BY month DESC, uf_sigla, product
    LIMIT 10;
""").df()
```

```python
#Fechar conexão
con.close()
```
